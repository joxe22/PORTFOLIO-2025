### Machine Learning

```python
## Defyning Variables
# y to predict
y = df['Price']
# X independent
X = df[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
```

```python
from sklearn.tree import DecisionTreeRegressor
## Creating the model
melb_model = DecisionTreeRegressor()
## Train the model
melb_model.fit(X, y)
```

```python
## Printing out
print("Prediction of The First Five House")
print(X.head())
print("The Real Price is:")
print(y.head())
print("Predictions Are:")
print(melb_model.predict(X.head()))
```

### MAE (MEAN ABSOLUTE ERROR)

VERIFICATION (ERRORS)

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

### VALIDATION


```python
## To evaluate the overfitting
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y)

melb_model = DecisionTreeRegressor()

melb_model.fit(train_X, train_y)

val_predictions = melb_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

Model failed with a price of 244,907 EUR

### Fuctions to fix the error

By creation a function where the model is the decisionTreeRegressor, but I need to change the max_leaf_node, predict X's values, and going to return MAE. This process is just to prove different shet's numbers in order to get better results

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

#### Documentation:

* With a leaf nodes of 5 sheet, we get an error of: 373,690
* With a leaf nodes of 50 sheet, we get an error of:259,584
* With a leaf nodes of 500 sheet, we get an error of:236,680 --> This one being the better result in DecisionTree Model
* With a leaf nodes of 5000 sheet, we get an error of:244,768
